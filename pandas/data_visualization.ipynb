## Questions

### 1. What are the 2 main approaches to use matplotlib?
- **Approach 1**: Using `pyplot` interface (state-based).
- **Approach 2**: Using Object-Oriented (OO) style.

**List of things you can do using one approach but not the other**:
- `pyplot`: Automatically manages the creation of figures and axes.
- **OO**: More control over axes, legends, and other plot elements; allows complex, multi-axes visualizations.

### 2. What is an ax?
- An "ax" (short for axes) refers to an instance of the `Axes` class in matplotlib. It represents a single plot within a figure and contains all the information and elements like data, ticks, labels, and gridlines.

### 3. Can you use seaborn in a matplotlib plot?
- Yes, seaborn is built on top of matplotlib. You can integrate seaborn plots into a matplotlib figure.

### 4. What is the difference between the visualization of categorical and numerical data?
- **Categorical data**: Represents distinct groups or categories (e.g., bar charts, pie charts).
- **Numerical data**: Represents quantitative values (e.g., histograms, line plots).

```python
import pandas as pd
btc = pd.read_csv('BTC-2019min.csv')
btc.head(20)
```

```python
import matplotlib.pyplot as plt
# Plot Bitcoin price evolution over 2019
btc['date'] = pd.to_datetime(btc['date'])  # Make sure the 'Date' column is in datetime format

plt.figure(figsize=(10, 6))
plt.plot(btc.loc[:,'date'], btc.loc[:,'close'], label='Bitcoin Price', color='orange')
# Add labels and title
plt.title('Bitcoin Price Evolution in 2019')
plt.xlabel('Date')
plt.ylabel('Price in USD')
plt.grid(True)

# Show the plot
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

### 2. Random Walk Simulation:
- Create a random walk simulation and visualize the distributions from the process.
- **Experiment with 1D and 2D random walks**.
- Investigate the theory behind random walks and compare your simulation results with theoretical expectations.

```python

import numpy as np
import matplotlib.pyplot as plt

# Number of steps in the random walk
num_steps = 10000

# Random walk: -1 (step backward) or +1 (step forward)
steps = np.random.choice([-1, 1], size=num_steps)

# Cumulative sum to simulate the random walk
random_walk = np.cumsum(steps)

# Plotting the random walk
plt.figure(figsize=(10, 6))
plt.plot(random_walk)
plt.title('1D Random Walk')
plt.xlabel('Steps')
plt.ylabel('Position')
plt.grid(True)
plt.show()

# Plotting the distribution of the final position (histogram)
plt.hist(random_walk, bins=30, density=True)
plt.title('Distribution of Final Positions in 1D Random Walk')
plt.xlabel('Position')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
# Number of steps in the random walk
num_steps = 1000

# Random steps for both x and y directions
steps_x = np.random.choice([-1, 1], size=num_steps)
steps_y = np.random.choice([-1, 1], size=num_steps)

# Cumulative sum for x and y directions
random_walk_x = np.cumsum(steps_x)
random_walk_y = np.cumsum(steps_y)

# Plotting the 2D random walk
plt.figure(figsize=(10, 6))
plt.plot(random_walk_x, random_walk_y)
plt.title('2D Random Walk')
plt.xlabel('X Position')
plt.ylabel('Y Position')
plt.grid(True)
plt.show()

# Scatter plot of the 2D random walk
plt.figure(figsize=(10, 6))
plt.scatter(random_walk_x, random_walk_y, c=np.arange(num_steps), cmap='viridis')
plt.title('2D Random Walk Path')
plt.xlabel('X Position')
plt.ylabel('Y Position')
plt.colorbar(label='Step Number')
plt.grid(True)
plt.show()
```
