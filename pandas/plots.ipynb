```python
import matplotlib.pyplot as plt

# Example Data
time = ['Jan', 'Feb', 'Mar', 'Apr', 'May']
price = [100, 105, 110, 115, 105]

plt.plot(time, price)
plt.title('Price Evolution Over Time')
plt.xlabel('Time')
plt.ylabel('Price')
plt.show()
```

```python
plt.fill_between(time, price, color='skyblue', alpha=0.4)
plt.plot(time, price, color='Slateblue', alpha=0.6)
plt.title('Price Evolution Over Time (Area Plot)')
plt.xlabel('Time')
plt.ylabel('Price')
plt.show()
```

```python
plt.bar(time, price)
plt.title('Price Evolution Over Time (Bar Chart)')
plt.xlabel('Time')
plt.ylabel('Price')
plt.show()
```

```python

# Example Data
time = ['Jan', 'Feb', 'Mar', 'Apr', 'May']
price = [100, 105, 110, 115, 120]

# Create a bar chart
bars = plt.bar(time, price)

# Add a red line at the top of each bar
for bar in bars:
    yval = bar.get_height()
    plt.plot([bar.get_x() + bar.get_width()/2, bar.get_x() + bar.get_width()/2], [0, yval], color='red', linestyle='-', linewidth=2)

# Add labels and title
plt.title('Price Evolution Over Time (Bar Chart with Red Lines)')
plt.xlabel('Time')
plt.ylabel('Price')

# Show the plot
plt.show()
```
